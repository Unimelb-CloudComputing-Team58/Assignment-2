## Define Grid Functions

```python
import geopandas as gpd
from shapely.geometry import Point, Polygon

class grid:
    """Reada in a shp file containing sa4 names and corresponding coordinates upon creation, 
    used to classify which sa4 from greater Melbourne area the tweets are from, and returns none otherwise
    
    Example usage: 

    - import grid
    path = '/Users/yanzu/Documents/projects/COMP90042-22S1-Project/tmp/aurin_data/shp_files/spatialise-SA4/shp/apm_sa4_2016_timeseries-.shp'
    - mel_grid = grid(path)
    - mel_grid.getSA4(Point(144.946457, -37.840935))
    
    """
    path = '/Users/yanzu/Documents/projects/git/Assignment-2/aurin_data/shp_files/spatialise-SA4/shp/apm_sa4_2016_timeseries-.shp'
    
    def __init__(self, shp_path = path):
        shp_frame = gpd.read_file(shp_path)
        self.SA4_DICT = dict(zip(shp_frame.sa42016na2, shp_frame.geometry))
        

    def getSA4(self, point):
        if point == '':
            return 'None'
        point = Point(point)
        if not point: 
            return 'None'
        for (sa4, polygon) in self.SA4_DICT.items():
            if point.within(polygon):
                return sa4
        return 'None'
    def getGrid(self):
        return self.SA4_DICT
```

```python
parks = gpd.read_file('tmp/project-data/greenspace_polygon_201908-.shp')
```

```python
# Get midpoint of a poly column 
from shapely import wkt
magrid = grid()
#grid3 = grid_sa3()
centre = []
SA4 = []
for index, row in named_parks.iterrows():
    mid = wkt.loads(row['geometry'].centroid.wkt)
    centre.append(mid)
    SA4.append(magrid.getSA4(mid))




named_parks['centroid'] = centre
named_parks['SA4'] = SA4
```
